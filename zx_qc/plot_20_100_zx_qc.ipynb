## 10 Qubits, 50 Depth

```python
df_dg = pd.read_pickle("/home/hinog/src/python3/pyzx_utils/zx_qc/data/pkl/20_100_zxdiagram.pkl")
df_qc = pd.read_pickle("/home/hinog/src/python3/pyzx_utils/zx_qc/data/pkl/20_100_qc.pkl")
df_qc1 = pd.read_pickle("/home/hinog/src/python3/pyzx_utils/zx_qc/data/pkl/20_100_qc1.pkl")
```

### num_gates in QC と zx の特徴量の関係性

```python
x_column = 'num_edge'  # x軸にしたい列名
y_column = 'num_gates'  # y軸にしたい列名

plt.scatter(df_dg[x_column], df_qc[y_column], s=10)
plt.xlabel(f'{x_column} in zx')
plt.ylabel(f'{y_column} in qc')
plt.show()

#相関係数の計算
correlation = df_dg[x_column].corr(df_qc[y_column])
print(correlation)
```
