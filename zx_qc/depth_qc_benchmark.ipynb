```python
import pyzx as zx
import random
import pandas as pd
import time
from collections import Counter
```

## pyzx depth-num_gates

```python
num_sample = 1000
qubit = [5, 10, 20, 40, 80, 100]
depth = [10, 20, 50, 80, 100, 200, 400]

qubit_depth_all = []
list = ["seed", "qubits", "depth", "num_gates"]

for i in range(num_sample):
    random.seed(i)
    start = time.time()
    for j in range(len(qubit)):
        for k in range(len(depth)):
            circ = zx.generate.cliffordT(qubits=qubit[j], depth=depth[k])
            num_gates = circ.num_vertices() - 4*circ.qubit_count()
            tmp_edge = circ.edge_set()
            flat = [item for sublist in tmp_edge for item in sublist]
            ct = Counter(flat)
            ct1 = {key: count for key, count in ct.items() if count >= 3}
            tmp1 = sum(ct1.values())
            tmp2 = 2 * len(ct1)
            num_2qgates = int((tmp1-tmp2)/2)
            num_gates = num_gates - num_2qgates
            tmp = [i, qubit[j], depth[k], num_gates]
            qubit_depth_all.append(tmp)
    end = time.time()
    ct = end - start
    print(f"seed{i} 経過時間:{ct}")

df = pd.DataFrame(qubit_depth_all, columns=list)
```
