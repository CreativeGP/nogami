```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

## 50 Qubits, 200 Depth

```python
df_dg = pd.read_pickle("/home/hinog/src/python3/pyzx_utils/zx_qc/data/pkl2/50_200_zxdiagram.pkl")
df_qc = pd.read_pickle("/home/hinog/src/python3/pyzx_utils/zx_qc/data/pkl2/50_200_qc.pkl")
df_qc1 = pd.read_pickle("/home/hinog/src/python3/pyzx_utils/zx_qc/data/pkl2/50_200_qc1.pkl")
```

### num_gates in QC と zx の特徴量の関係性

```python
x_column = 'num_edge'  # x軸にしたい列名
y_column = 'num_gates'  # y軸にしたい列名


plt.scatter(df_dg[x_column], df_qc1[y_column], s=10)
plt.xlabel(f'{x_column} in zx')
plt.ylabel(f'{y_column} in qc')

# 1次関数の近似
coefficients = np.polyfit(df_dg[x_column], df_qc1[y_column], 1)
polynomial = np.poly1d(coefficients)  
ys = polynomial(df_dg[x_column])
plt.plot(df_dg[x_column], ys, color='red')
plt.show()

#相関係数の計算
correlation = df_dg[x_column].corr(df_qc1[y_column])
print("相関係数",correlation)

# 近似曲線の式
equation = f"y = {polynomial[1]:.2f}x + {polynomial[0]:.2f}"
print("近似曲線の式",equation)
```
